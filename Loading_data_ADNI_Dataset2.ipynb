```python
import csv
from scipy import ndimage
from scipy import linalg
import numpy as np
import nibabel as nib
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
from operator import itemgetter
import pickle
```

```python
#loading data


cwd = os.getcwd()#return current working directory
patient_id = os.listdir('ADNI') #create a list that encompasses all the images we downloaded
filename = glob.glob('data/' + '*/data_batch*')

patient_num = 2084
datapath = [] #[patient_id, path]
dataimage = []#[patient_id, image_matrix_normalized]


#import csv files
datacsv = np.genfromtxt('ADNI1_Complete_1Yr_1.5T_1_07_2019.csv', delimiter = ',', encoding = None, dtype = None)


datacsv = np.delete(datacsv, 0, 1)
datacsv = np.delete(datacsv, np.s_[4:12], 1)
datacsv = np.delete(datacsv, 0, 0)

#rename csv group into numbers
for i in range(datacsv.shape[0]):
    if datacsv[i][1] == '"MCI"':
        datacsv[i][1] = 0
    elif datacsv[i][1] == '"AD"':
        datacsv[i][1] = 1
    elif datacsv[i][1] == '"CN"':
        datacsv[i][1] = 2

pat_id = datacsv[0][0]
flag_0 = 0
datacsv_new = datacsv.copy()#[patient_id, group, sex, age]

deleted_line = 0

#filter csv
for i in range(datacsv.shape[0]):
    if (pat_id == datacsv[i][0]) & (flag_0 >= 1):
        datacsv_new = np.delete(datacsv_new, i - deleted_line, 0)
        deleted_line = deleted_line + 1
    elif (pat_id == datacsv[i][0]) & (flag_0 == 0):
        flag_0 = flag_0 + 1
    elif (pat_id != datacsv[i][0]):
        flag_0 = 1
        pat_id = datacsv[i][0]
        
#datapath part
for patient in patient_id:
    flag = 0
    print("haha")
    for root, dirs, files in os.walk(cwd + "/ADNI/" + patient):         
        for file in files:
            if file.endswith(".nii"):
                if flag < 1:
                    #print([patient, os.path.join(root, file)])####
                    datapath.append([patient, os.path.join(root, file)])
                    flag = flag + 1

                    
#dataimage part
tar_dim = [256, 256, 160]

for path in datapath:
    print("greatjob")
    mri = nib.load(path[1]).get_data()
    mri = (mri - mri.min())/(mri.max() - mri.min())
    mri = mri - mri.mean()
    tar_mri = ndimage.zoom(mri,[tar_dim[0]/mri.shape[0],tar_dim[1]/mri.shape[1], tar_dim[2]/mri.shape[2]], order = 1)

    #print(mri)
    #print(mri.mean())
    #print(mri.var())
    #print(tar_mri.shape)
    #plt.imshow(mri[125,:,:],'gray')
    #plt.show()
    #print([path[0],tar_mri.shape])
    
    dataimage.append([path[0],tar_mri])
#print(dataimage)
#print(len(dataimage))
#print(len(dataimage[0]))
dataimage = sorted(dataimage, key=itemgetter(0), reverse=True)

#for image in dataimage:
#    print(image[0])
#print(datacsv_new)
#print(len(datacsv_new))
#print(len(datacsv_new[0]))
#dataiamge_arr = np.asarray(dataimage)
#datacsv_new_arr = np.asarray(datacsv_new)
#np.save('dataimage_arr', dataiamge_arr)
#np.save('datacsv_new_arr', datacsv_new_arr)
```
